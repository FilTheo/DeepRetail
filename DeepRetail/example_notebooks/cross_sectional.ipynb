#### Reads

```python
import pandas as pd
import numpy as np
from DeepRetail.reconciliation.cross_sectional import CHieF, CrossSectionalReconciler
```

```python
df = pd.read_csv('/home/filtheo/DeepRetail/DeepRetail/data/data_folder/m5_day.csv', index_col=0)
df.head()
```

```python
# Go monthly
df.columns = pd.to_datetime(df.columns)
df = df.resample('M', axis = 1).sum()
```

#### Define Variables

```python
# Take just a sample from the df 
# A complete state.

# Filter the original dataframe
sample_df = df[(df.index.str.contains('WI_1'))| (df.index.str.contains('WI_2'))]
```

```python
# Initialize arguments

# For building the hierarchical index:
current = ['cat', 'catnum', 'itemnum', 'state', 'storenum']
correct = ['itemnum', 'catnum', 'cat', 'storenum', 'state']
splitter = "_"
total = False 
```

```python
# For model initialization
freq = 'M'
h = 12
```

```python
# For forecasting
models = ['ETS']
```

#### Example 1: CHieF & No Holdout

```python
holdout = False
```

```python
# Define CHieF
chief = CHieF(bottom_level_freq = freq, h=h, holdout = holdout)
```

```python
# Fit CHieF -> builds the hierarchical format
chief.fit(df = sample_df, current_format=current, corrected_format=correct, splitter=splitter, add_total = total, format = 'pivoted')
```

```python
# The hierarchical index
chief.hierarchical_format
```

```python
# the smatrix
chief.S_mat
```

```python
# The new hierarchical df with all hierarchical levels
chief.cross_sectional_df
```

```python
# Get all base forecasts
base_forecasts_1 = chief.predict(models = models)
```

```python
base_forecasts_1.head()
```

```python
# Define the reconciliation method
method = 'struc'
```

```python
reconciled_forecasts_1 = chief.reconcile(method = method)
```

```python
reconciled_forecasts_1.head()
```

#### Example 2: Use Holdout

```python
holdout = True
cv = 1 # one period for holdout
```

```python
# Define CHieF
chief = CHieF(bottom_level_freq = freq, h=h, cv =cv, holdout = holdout)

# Fit
chief.fit(df = sample_df, current_format=current, corrected_format=correct, splitter=splitter, add_total = total, format = 'pivoted')

# Get Base Forecasts
base_fc = chief.predict(models = models)

# Reconcile
method = 'mse' # use residuals this time 
reconciled_fc = chief.reconcile(method)
```

```python
reconciled_fc.head()
```

#### Example 3: Use the Reconciler Only with Cross-Validation

```python
from DeepRetail.forecasting.statistical import StatisticalForecaster
```

```python
holdout = True
cv = 2
```

```python
# Get the the hierarchical df and the s matrix
hierarchical_df = chief.cross_sectional_df
s_mat = chief.S_mat
```

```python
# Define a forecaster
forecaster = StatisticalForecaster(models = models, freq = freq)

# Fit the forecaster
forecaster.fit(hierarchical_df, format = 'pivoted')

# predict
base_forecasts = forecaster.predict(h = h, cv = cv, holdout = holdout)
```

```python
base_forecasts.head()
```

```python
# Get the residuals
residuals = forecaster.calculate_residuals()
```

```python
# Define the reconciler
reconciler = CrossSectionalReconciler(bottom_level_freq=freq, h = h, holdout = holdout, cv = cv)
```

```python
# fit the reconciler
reconciler.fit(df = base_forecasts, s_mat = s_mat)
```

```python
# reconcile
method = 'var' # variance scalling
reconciled_forecasts = reconciler.reconcile(method = method, residual_df = residuals)
```
